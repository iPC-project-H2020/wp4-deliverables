```python
df = pd.read_csv(folder_work+'_OFTENreportSummary.txt',sep='\t')
display(df)
df1 = df.loc[(df.SCORE>0.05)&(df.PVAL==0)]
display(df1)
df_ICA = pd.read_csv(folder+prefix+'_ica_S.xls',sep='\t')
df_ICA = df_ICA[df_ICA.columns[:-1]]

gr_combined = nx.Graph()
with open(folder_work+'_network_genes.txt','w') as fout:
    fout.write('GENE\tCOMPONENT\n')
    for r in df1.iterrows():
        print(r[1]['LABEL'])
        fn = prefix+'_'+r[1]['LABEL']+'_'+r[1]['TYPE'].lower()+'.xgmml'
        with open(folder_work+fn,'rb') as f:
            gr = XGMMLReader(f)
            for gn in gr.nodes():
                dfg = df_ICA.loc[df_ICA.PROBE==gn]
                vec = np.abs(dfg[dfg.columns[1:]].to_numpy())
                k = np.argmax(vec)
                comp = dfg.columns[k+1]
                fout.write(gn+'\t'+comp+'\n')
            gr_combined = nx.compose(gr_combined,gr)
print(gr_combined)
with open(folder_work+'_combined.sif','w') as f:
    for e in gr_combined.edges():
        f.write(e[0]+'\tpp\t'+e[1]+'\n')
        
```
